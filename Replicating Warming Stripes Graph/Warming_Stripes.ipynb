```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## get data from file
data = np.loadtxt("HadCRUT.4.6.0.0.annual_ns_avg 2.txt", skiprows=2)
```

```python
data[0]
```

```python
temp_data = data[:,11]

stacked_temp = np.stack((temp_data, temp_data))

vmin = 5.9
vmax = 11
## plotting
plt.figure(figsize=(15,18))

img = plt.imshow(stacked_temp, cmap='RdBu_r', aspect=40, )

plt.gca().set_axis_off()
plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0,
            hspace = 0, wspace = 0)
plt.margins(0,0)

plt.gca().xaxis.set_major_locator(plt.NullLocator())
plt.gca().yaxis.set_major_locator(plt.NullLocator())

plt.savefig("stripes.png", bbox_inches = 'tight', pad_inches = 0, dpi=400)
```
